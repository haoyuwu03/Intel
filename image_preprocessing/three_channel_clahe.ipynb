```python
#import neccessary files
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def clahe_image(image, color, clip_limit, BGR=True):

    #converts the image into RGB if not alred in RGB format
    if (BGR == True):
        old_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    elif (BGR == False):
        old_image = image

    #splits image into its rgb channels
    R,G,B = cv2.split(old_image)

    #defines the clahe cv2 function
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=(8,8))


    if (color == 'red'):
        R = clahe.apply(R)
        return(R) #remove the return if you want a full image returned
    elif (color == 'green'):
        G = clahe.apply(G)
        return(G) #remove the return if you want a full image returned
    elif (color == 'blue'):
        B = clahe.apply(B)
        return(B) #remove the return if you want a full image returned

    '''
    if you want function to return a combined image:

    new_image = cv2.merge((R, G, B))
    rerurn(new_image)
    ''' 
```

```python
#CLAHE of red, green and blue channels
temp_image = cv2.imread('D:\d_retinopathy_training\diabetic_rectinopathy-0-00cc2b75cddd.png')
colors = ['green', 'red', 'blue']

g = clahe_image(temp_image, colors[0], clip_limit = 2)
r = clahe_image(temp_image, colors[1], clip_limit = 2)
b = clahe_image(temp_image, colors[2], clip_limit = 1)

new_image = cv2.merge((r,g,b))
plt.imshow(new_image)
```
