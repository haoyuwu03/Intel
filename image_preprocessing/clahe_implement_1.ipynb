```python
#import neccessary files
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("D:\d_retinopathy_training\diabetic_rectinopathy-0-00cc2b75cddd.png", 0)
clahe = cv2.createCLAHE(clipLimit = 8.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
```

```python
print("CLAHE Adjusted Fundus Image")
plt.imshow(cl1)
```

```python
print("Original Image")
plt.imshow(img)
```

```python
#saves CLAHE adjusted file.
cv2.imwrite("D:\d_retinopathy_training\Test.png",cl1)
```
