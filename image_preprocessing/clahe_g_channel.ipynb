```python
#import neccessary files
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
from PIL import Image
```

```python
#CLAHE v2. 

'''
Code is designed from "Color Retinal Image Enhancement using CLAHE" research paper.
'''

#Prereqs.
old_image = cv2.imread('D:\d_retinopathy_training\diabetic_rectinopathy-0-00cc2b75cddd.png')

#Step 1: Channel Splitting of the BGR Image.
old_image = cv2.cvtColor(old_image, cv2.COLOR_BGR2RGB)
R,G,B = cv2.split(old_image)

#Step 2: Apply ClAHE on the "g" channel of the image
clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(8,8))
clahe_g_channel = clahe.apply(G)

#Step 3: Merge Image Channels
clahe_image1 = cv2.merge((R, clahe_g_channel, B))
print('New image with CLAHE implemented on the "g" channel:')
plt.imshow(clahe_image1)
```
